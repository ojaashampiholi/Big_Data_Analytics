```python
def calculateSentimentScore(df, col):
    sentScore = 0
    for i in range(df.shape[0]):
        word = df[col][i]
        word = str(word)
        sentiment2 = TextBlob(word)
        sentScore += sentiment2.sentiment.polarity
    avgScore = sentScore/ df.shape[0]
    return avgScore
    
avgScoreTwitter = calculateSentimentScore(dfTwitterOutput, "word")
print("The sentiment score for Twitter data is", avgScoreTwitter)

avgScoreStorm = calculateSentimentScore(dfStormOutput, "word")
print("The sentiment score for the Storm data is", avgScoreStorm)
```

```python
avgScoreStorm = calculateSentimentScore(dfStormInput, "EVENT_NARRATIVE")
print("The sentiment score for the input data of storm is", avgScoreStorm)

avgTwitterStorm = calculateSentimentScore(dfTwitterInput, "text_no_rt")
print("The sentiment score for the data is", avgTwitterStorm)
```
